```python
# 21BDS0325
import numpy as np
import pandas as pd

# 2-way contingency table for Categorical-Categorical variables
cat_cat_table = pd.crosstab(df['Latitude'], df['Longitude'])
print(cat_cat_table)

# For Numerical-Numerical (using Pearson's correlation for numeric values like Latitude and Longitude)
correlation = df['Latitude'].corr(df['Longitude'])
print("Correlation between Age and Fare:", correlation)

# Categorical-Numerical: aggregate numerical variables by categorical values
agg_fare_by_income = df.groupby('Latitude')['Longitude'].mean()
print("Mean Fare by Latitude:\n", agg_fare_by_income)
```

```python
# 21BDS0325
three_way_table = pd.crosstab([df['Latitude'], df['Longitude']], df['Restaurant_ID'])
print(three_way_table)
```

```python
#21BDS0325
from scipy.stats import chi2_contingency

# Row profile (each value divided by the row total)
row_profile = cat_cat_table.div(cat_cat_table.sum(axis=1), axis=0)
print("Row Profile:\n", row_profile)

# Column profile (each value divided by the column total)
column_profile = cat_cat_table.div(cat_cat_table.sum(axis=0), axis=1)
print("Column Profile:\n", column_profile)

# Relative frequency (each value divided by the total sum of the table)
relative_frequency = cat_cat_table / cat_cat_table.values.sum()
print("Relative Frequency:\n", relative_frequency)

# Chi-square test for independence between Survived and Pclass
chi_square_result = chi2_contingency(cat_cat_table)
print("Chi-square Test:\n", chi_square_result)
```

# K-Means Clustering

```python
#21BDS0325

# Import necessary libraries
from google.colab import files
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import seaborn as sns


# Display dimensions of the dataset
print("Dimensions of the dataset:", df.shape)

# Display summary of the dataset
print("\nSummary of the dataset:")
print(df.describe(include='all'))

# Step 2: Extract the relevant columns (Latitude and Longitude)
X = df[['Latitude', 'Longitude']]
print("Extracted columns: Latitude and Longitude")
print(X.head())

# Step 3: Use the Elbow Method to find the optimal number of clusters
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=123)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
print("WCSS for each number of clusters (1 to 10):")
print(wcss)

# Step 4: Plot the Elbow Method
plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), wcss, marker='o', linestyle='-', color='b')
plt.title("Elbow Method for Optimal Clusters")
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.show()

# Step 5: Based on the Elbow method, set the number of clusters
optimal_clusters = 5  # Assuming the elbow suggests 5 clusters; adjust based on the plot
kmeans = KMeans(n_clusters=optimal_clusters, init='k-means++', random_state=123)
y_kmeans = kmeans.fit_predict(X)
print("K-Means clustering applied with", optimal_clusters, "clusters")

# Step 6: Add the cluster assignments to the original dataframe for analysis
df['Cluster'] = y_kmeans

# Step 7: Cluster visualization
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Latitude', y='Longitude', data=df, hue='Cluster', palette="viridis", s=60)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='red', label='Centroids')
plt.title("Clusters of Restaraunts")
plt.xlabel("Latitude")
plt.ylabel("Longitude")
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.spatial.distance import pdist

df1=df.head(10)
# Step 2: Extract the relevant columns (Latitude and Latitude)
X = df1[['Latitude', 'Latitude']]
print("Extracted columns: Latitude and Latitude")
#print(X.head())

# Step 3: Define a function to perform hierarchical clustering for different distance measures
def perform_hierarchical_clustering(data, distance_method):
    # Compute the distance matrix
    distance_matrix = pdist(data, metric=distance_method)

    # Perform hierarchical clustering using Ward's method for linkage
    linkage_matrix = linkage(distance_matrix, method='ward')

    # Plot the dendrogram
    plt.figure(figsize=(10, 5))
    dendrogram(linkage_matrix)
    plt.title(f"Hierarchical Clustering - {distance_method} distance")
    plt.xlabel("Data Points")
    plt.ylabel("Height")
    plt.show()

# Step 4: Perform hierarchical clustering with different distance measures

# Euclidean Distance
perform_hierarchical_clustering(X, "euclidean")

# Manhattan Distance
perform_hierarchical_clustering(X, "cityblock")

# Maximum Distance
perform_hierarchical_clustering(X, "chebyshev")

# Canberra Distance
perform_hierarchical_clustering(X, "canberra")

# Binary Distance (for binary attributes, this may not be suitable for continuous data)
perform_hierarchical_clustering(X, "hamming")

# Minkowski Distance (default p=1; equivalent to Euclidean for p=2)
perform_hierarchical_clustering(X, "minkowski")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Step 2: Select relevant numerical columns for PCA
numeric_data = df.select_dtypes(include=[np.number])
print("Selected numeric columns for PCA:")
print(numeric_data.head())

# Step 3: Standardize the data
scaler = StandardScaler()
numeric_data_scaled = scaler.fit_transform(numeric_data)
print("Data standardized")

# Step 4: Perform PCA
pca = PCA()
pca_result = pca.fit(numeric_data_scaled)
print("PCA performed")

# Step 5: Summary of PCA results (variance explained by each component)
explained_variance = pca.explained_variance_ratio_
print("Explained variance by each principal component:")
print(explained_variance)

# Step 6: Visualize the variance explained by each principal component
plt.figure(figsize=(10, 5))
plt.bar(range(1, len(explained_variance) + 1), explained_variance, alpha=0.7, color="b")
plt.xlabel("Principal Components")
plt.ylabel("Variance Explained")
plt.title("Variance Explained by Each Principal Component")
plt.show()

# Step 7: Biplot visualization
pca = PCA(n_components=2)  # Only using the first two components for visualization
numeric_data_pca = pca.fit_transform(numeric_data_scaled)

plt.figure(figsize=(10, 7))
sns.scatterplot(x=numeric_data_pca[:, 0], y=numeric_data_pca[:, 1], palette="viridis")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA - Biplot of Credit Card Data")

# Plot vectors for original features
for i, feature in enumerate(numeric_data.columns):
    plt.arrow(0, 0, pca.components_[0, i], pca.components_[1, i],
              color="r", alpha=0.5)
    plt.text(pca.components_[0, i] * 1.15, pca.components_[1, i] * 1.15,
             feature, color="g", ha="center", va="center")
plt.show()

# Step 8: Contribution of variables to the principal components
plt.figure(figsize=(10, 5))
plt.bar(range(1, len(pca.explained_variance_ratio_) + 1), pca.explained_variance_ratio_, color="lightblue")
plt.xlabel("Principal Components")
plt.ylabel("Explained Variance Ratio")
plt.title("Variable Contributions to Principal Components")
plt.show()
```
